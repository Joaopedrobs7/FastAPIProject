```python
from transformers import pipeline

# Pipeline zero-shot multil√≠ngue
#classificador = pipeline("zero-shot-classification",model="neuralmind/bert-base-portuguese-cased")
classificador = pipeline(
    "zero-shot-classification",
    model="joeddav/xlm-roberta-large-xnli"
)

labels = [
    "Produtivo: Emails que requerem uma a√ß√£o ou resposta espec√≠fica (ex.: solicita√ß√µes de suporte t√©cnico, atualiza√ß√£o sobre casos em aberto, d√∫vidas sobre o sistema)",
    "Improdutivo: Emails que n√£o necessitam de uma a√ß√£o imediata (ex.: mensagens de felicita√ß√µes, agradecimentos)"
]

# Escolha um bloco

predicoes = []

for texto, rotulo_real in emails_maiores:
    resultado = classificador(texto, candidate_labels=labels)
    label_predita = resultado['labels'][0]
    predicoes.append((texto, rotulo_real, label_predita, resultado['scores'][0]))

# Calcular acur√°cia
acertos = sum(1 for _, rotulo_real, label_predita, _ in predicoes if rotulo_real in label_predita)
acuracia = acertos / len(predicoes)
print(f"Acur√°cia: {acuracia*100:.2f}%")

# Detalhes
rows = 0
for texto, rotulo_real, label_predita, score in predicoes:
    rows+=1
    print(f"\nTexto: {texto}")
    print(f"Rotulo real: {rotulo_real}")
    print(f"Predito: {label_predita} ({score:.2f})")
    print(f"texto n:{rows}")
```

```python
%pip install datasets

from datasets import Dataset

# Converter lista de emails para Dataset
dataset = Dataset.from_list([
    {"text": t, "label": 1 if l=="Produtivo" else 0} for t, l in todos_emails
])

# Fazer train/test split
dataset_split = dataset.train_test_split(test_size=0.2, seed=42)
train_dataset = dataset_split['train']
test_dataset = dataset_split['test']

print("Exemplos de treino:", len(train_dataset))
print("Exemplos de teste:", len(test_dataset))
```

```python
%pip install setfit transformers sentence-transformers
```

```python
from setfit import SetFitModel, SetFitTrainer
from sklearn.metrics import accuracy_score

model_id = "neuralmind/bert-base-portuguese-cased"
model = SetFitModel.from_pretrained(model_id)
```

```python
import os
os.environ["WANDB_DISABLED"] = "true"
trainer = SetFitTrainer(
    model=model,
    train_dataset=train_dataset,
    eval_dataset=test_dataset,
    metric=accuracy_score,
    batch_size=8,
    num_iterations=20  # ajuste se quiser mais precis√£o
)

trainer.train()
```

```python
metrics = trainer.evaluate()
print("Acur√°cia no teste:", metrics)
```

```python
# Predi√ß√£o de r√≥tulo
preds = model.predict([t[0] for t in todos_emails])

# Predi√ß√£o de probabilidades
probs = model.predict_proba([t[0] for t in todos_emails])

# Inicializa contagem de acertos
acertos = 0

# Mostrar resultados
for (email, rotulo_real), pred, prob in zip(todos_emails, preds, probs):
    label = "Produtivo" if pred == 1 else "Improdutivo"

    # Contar acertos
    if label == rotulo_real:
        acertos += 1

    # Converter tensores para float
    prob_produtivo = float(prob[1])
    prob_improdutivo = float(prob[0])

    print(f"üì© {email} => {label}")
    print(f"Rotulo real: {rotulo_real}")
    print(f"Produtivo: {round(prob_produtivo*100,2)}% | Improdutivo: {round(prob_improdutivo*100,2)}%\n")

# Calcular e imprimir acur√°cia
acuracia = acertos / len(todos_emails)
print(f"‚úÖ Acur√°cia do modelo: {acuracia*100:.2f}%")
```
